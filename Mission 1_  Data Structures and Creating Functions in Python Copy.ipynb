```python
#  Code Here Exercise 6
# Define a function that takes two numbers and returns their difference
def find_difference(num1, num2):
    return num1 - num2

# Call the function with different numbers and print the results
result1 = find_difference(10, 4)
result2 = find_difference(25, 8)
result3 = find_difference(100, 50)

# Display the results
print("Difference 1:", result1)
print("Difference 2:", result2)
print("Difference 3:", result3)
```
